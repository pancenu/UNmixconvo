## 1. Bloque de funciones necesarias para busqueda

1. descargar_carpeta_github

```python
import requests
import os

# Función para establecer las palabras clave a buscar

def establecer_palabras_clave() -> list:
    """
    Establece la lista de palabras clave basado en la entrada del usuario.

    Returns:
        Lista de palabras clave.
    """
    try:
        numero_palabras_clave = input("Digita el número de palabras clave a buscar: ")
        numero_palabras_clave = int(numero_palabras_clave)

        if numero_palabras_clave <= 0:
            print("Debes digitar un número entero positivo.")
            exit(1)

    except ValueError:
        print("Error: Debes digitar un número entero válido.")
        exit(1)

    palabras_clave = []
    for i in range(numero_palabras_clave):
        palabra_clave = input(f"Digita la palabra clave #{i+1}: ").strip()
        if palabra_clave:  # Check if input is not empty
            palabras_clave.append(palabra_clave)
        else:
            print("Error: No puedes ingresar una palabra clave vacía.")
            exit(1)
    return palabras_clave

# Función para descargar los archivos de texto directamente de github

def descargar_carpeta_github(carpeta):
  """
  Descarga el contenido de la una carpeta del repositorio,
  podría descargar la que contiene los pdf o la que contiene los archivos de texto generados
  el proceso original tardo cerca de tres horas, descargando toma de dos a tres minutos
  """
  # Crear carpeta para guardar txt's
  os.makedirs(carpeta, exist_ok=True)

  response = requests.get("https://api.github.com/repos/pancenu/UNmixconvo/contents/" + carpeta)
  for item in response.json():
      archivo = item["name"]
      url = "https://github.com/pancenu/UNmixconvo/blob/main/"+ carpeta +"/" +  archivo
      response = requests.get(url)
      if response.status_code == 200:
          with open(os.path.join(carpeta, archivo), "wb") as f:
              f.write(response.content)
# Para mostrar los nombres de los  archivos a medida que se descargan
# descomentar, eliminar el caracter "#", de la siguiente línea
          #print(f"✅ Descargado: {archivo}")
      else:
          print(f"❌ Error al descargar {archivo} (Código: {response.status_code})")


# Función para generar la lista de archivos (cargos) que contienen las palabras clave buscadas

def buscar_palabras_clave(palabras_clave):
  list_txt = []
  for txt_file in os.listdir("texto_extraido"):
      if txt_file.endswith(".txt"):
          txt_path = os.path.join("texto_extraido", txt_file)
          print(f"\nProcesando: {txt_file}")

          # Abre el archivo en modo lectura ('r')
          with open(txt_path, "r") as f:
              # Lee todo el contenido del archivo
              texto = f.read()
              # Buscar palabras clave
              coincidencias = [palabra for palabra in palabras_clave
                              if palabra.lower() in texto.lower()]
              if coincidencias:
                  print(f"✅ Coincidencias en {txt_file}: {', '.join(coincidencias)}")
                  list_txt.append(txt_file[:-4])
              else:
                  print(f"⚠️ No se encontraron coincidencias en {txt_file}")
  return list_txt

def listar_resultado(lista_resultado):
    url_base_down = "https://concursomixto.unal.edu.co/unalmeritos/files/v1.0/file/static/announcements/"
    print("\n Enlaces a los archivos de la lista: \n")
    for archvivo in lista_resultado:
        url_archivo = url_base_down + archvivo
        print(url_archivo)


```

## Consultar la página de la convocatoria y extraer las url (enlaces) de cada uno de los archivos pdf (cargos, limitado a la convocatoria Abierta).

```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.options import Options
import time

# Configuración de Firefox
options = Options()
options.add_argument('--headless')

driver = webdriver.Firefox(service=Service(), options=options)

# Navegar a la página
driver.get("https://concursomixto.unal.edu.co/#/announcement")
time.sleep(5)

pdf_links = []

while True:
    # Extraer PDFs
    pdf_elements = driver.find_elements(By.XPATH, "//a[contains(@href, 'Ab.pdf')]")
    for element in pdf_elements:
        pdf_url = element.get_attribute("href")
        if pdf_url not in pdf_links:
            pdf_links.append(pdf_url)
            print("PDF encontrado:", pdf_url)

    # Paginación (ajusta el XPath según HTML)
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH,
                "//li[@class='waves-effect']/a[i[@class='material-icons' and contains(text(), 'chevron_right')]]"
            ))
        )
        if "disabled" in next_button.find_element(By.XPATH, "./parent::li").get_attribute("class"):
            print("Fin de la paginación.")
            break
        next_button.click()
        time.sleep(3)
    except Exception as e:
        print("Error al avanzar:", e)
        break

print("\nTotal de PDFs:", len(pdf_links))
driver.quit()
```

## Descarga cada uno de los archivos de la lista creada.

```python
import os
import requests
from urllib.parse import unquote, urlparse

# Crear carpeta para guardar PDFs
os.makedirs("pdfs_descargados", exist_ok=True)

# Lista de URLs de PDFs (obtenida previamente con Selenium)

def obtener_nombre_archivo(url):
    """Extrae el nombre del archivo desde la URL, decodificando caracteres especiales."""
    url_path = unquote(urlparse(url).path)  # Decodifica %20, %C3%A9, etc.
    return os.path.basename(url_path) or "documento_sin_nombre.pdf"  # Usar nombre por defecto si no hay

# Descargar cada PDF conservando su nombre original
for url in pdf_links:
    try:
        nombre_archivo = obtener_nombre_archivo(url)
        pdf_path = os.path.join("pdfs_descargados", nombre_archivo)

        # Evitar descargas duplicadas
        if os.path.exists(pdf_path):
            print(f"⚠️ Ya existe: {nombre_archivo} (omitido)")
            continue

        response = requests.get(url, stream=True)
        response.raise_for_status()  # Verificar errores HTTP

        with open(pdf_path, "wb") as f:
            for chunk in response.iter_content(1024):
                f.write(chunk)
        print(f"✅ Descargado: {nombre_archivo}")

    except Exception as e:
        print(f"❌ Error al descargar {url}: {str(e)}")
```

## Función para la extracción del texto de archivos

```python
# función que extrae texto de documentos "escaneados"

import pytesseract
from pdf2image import convert_from_path
import os

def pdf_imagen_a_texto(pdf_path):
    """Convierte un PDF de imágenes a texto usando OCR."""
    texto_extraido = ""

    # Convertir PDF a imágenes (una por página)
    imagenes = convert_from_path(pdf_path, dpi=300)  # Ajusta DPI para calidad

    for i, img in enumerate(imagenes):
        # Configuración de Tesseract (idioma español)
        texto_pagina = pytesseract.image_to_string(img, lang='spa')
        texto_extraido += f"\n--- Página {i+1} ---\n{texto_pagina}"

    return texto_extraido
```

## Extracción y almacenamiento como archivos de texto.

```python
# Crear carpeta para guardar texto extraido
# se usará este texto para búsqueda.

os.makedirs("texto_extraido", exist_ok=True)

# Procesar cada PDF
for pdf_file in os.listdir("pdfs_descargados"):
    if pdf_file.endswith(".pdf"):
        if os.path.exists(f"texto_extraido/{pdf_file}.txt"):
            print(f"⚠️ Ya existe: {pdf_file} (omitido)")
        else:
          pdf_path = os.path.join("pdfs_descargados", pdf_file)
          print(f"\nProcesando: {pdf_file}")

          try:
            # Extraer texto con OCR
            texto = pdf_imagen_a_texto(pdf_path)

            # Guardar texto extraído para referencia
            with open(f"texto_extraido/{pdf_file}.txt", "w") as f:
              f.write(texto)

            if False:
              # Buscar palabras clave
              # Originalmente se hace la búsqueda a medida que pasa a texto
              coincidencias = [palabra for palabra in palabras_clave
                              if palabra.lower() in texto.lower()]

              if coincidencias:
                  print(f"✅ Coincidencias en {pdf_file}: {', '.join(coincidencias)}")

              else:
                  print(f"⚠️ No se encontraron coincidencias en {pdf_file}")
          except Exception as e:
              print(f"❌ Error al procesar {pdf_file}: {str(e)}")
```
